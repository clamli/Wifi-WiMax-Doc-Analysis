```python
from xlrd import open_workbook
import requests
```

### Get Files of 802_16_Documents

```python
links = getUrl("802_16_Documents.xls")
total = len(links)
for idx, link in zip(range(total), links):
    print("%.2f%%"%((idx+1)*100/total), end='\r')
    getFileFromUrl("80216Doc/", link)
```

### Get Files of 802_16_Completed_Projects

```python
links = getUrl("802_16_Completed_Projects.xls")
total = len(links)
for idx, link in zip(range(total), links):
    print("%.2f%%"%((idx+1)*100/total), end='\r')
    if idx*100/total <= 87.18:
        continue
    getFileFromUrl("80216CompProj/", link)
```

### Util functions

```python
def getUrl(filename):
    book = open_workbook(filename, on_demand=True)
    book.sheet_names()
    links = []
    for name in book.sheet_names():
        sheet = book.sheet_by_name(name)
        for idx, cell in zip(range(sheet.ncols), sheet.row(0)):
            if 'File Link' in cell.value:
                break
        for cell in sheet.col(idx)[1:]:
            links.append(cell.value)
    return links
```

```python
def getFileFromUrl(path, url):
    filename = url[url.rfind('/')+1:]
    response = requests.get(url, allow_redirects=True)
    with open(path + filename, 'wb') as file:
        file.write(response.content)
```
